```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

```

```python
df=pd.read_csv('Titanic.csv')
df
```

```python
df['age'].plot(kind='box')
plt.title('boxplot')
plt.show
```

```python
plt.scatter(df['age'],df['fare'])
plt.xlabel('age')
plt.ylabel('fare')
plt.title('scatter plot')
plt.show
```

```python
q1=df['age'].quantile(0.25)
q1
q3=df['age'].quantile(0.75)
q3
iqr=q3-q1
lowerbound=q1+(iqr*1.5)
lowerbound
upperbound=q3+(iqr*1.5)
upperbound
```

```python
age_arr=df['age']
outliers=age_arr[(age_arr<=lowerbound)|(age_arr>=upperbound)]
outliers
```

```python
lowerindex=df[df['age']<lowerbound].index
df.drop(lowerindex,inplace=True)
df.info()
```

```python
fare_arr=df['fare']
lowercap=np.percentile(fare_arr,99)
uppercap=np.percentile(fare_arr,1)
```

```python
for i in df['fare']:
    if i<lowerbound:
        df['age']= df['age'].replace(i,lowercap)
    if i>upperbound:
        df['age']= df['age'].replace(i,uppercap)
df.info()
    
```

```python
m=df['age'].mean()
print(m)
for i in df['age']:
    if i<lowerbound or i>upperbound:
        df['age']= df['age'].replace(i,m)
        
    
```

```python
m=df['age'].median()
print(m)
for i in df['age']:
    if i<lowerbound or i>upperbound:
        df['age']= df['age'].replace(i,m)
        
```
